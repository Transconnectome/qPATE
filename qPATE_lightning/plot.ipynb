```python
import os 
import glob
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
import seaborn as sns
```

# Epsilon vs. ACC

```python
## settings 
seed_dir = glob.glob('/Users/wangheehwan/Desktop/qPATE_lightning/result/*')
cutoff_epoch = 20
```

```python
classical_eps_all = []
classical_results_all = []
for seed in seed_dir: 
    PATE_result_dir = os.path.join(seed, 'cPATE_epsilon.csv')
    if os.path.exists(PATE_result_dir):
        
        PATE_result = pd.read_csv(PATE_result_dir)
        #classical_seed = []
        classical_eps = [] 
        classical_results = [] 
        for i, k in enumerate(PATE_result.columns): 
            eps = float(float(k.replace(' - test_student_acc', '').split('_')[2].replace('e','')))
            test_student_acc = PATE_result[k].values[cutoff_epoch-1]
            classical_eps.append(eps)
            classical_results.append(test_student_acc)
            #classical_seed.append(float(k.replace(' - test_student_acc', '').split('_')[0].replace('s',''))) 
            

        classical_eps = np.array(classical_eps)
        classical_results = np.array(classical_results)
        #classical_seed = np.array(classical_seed) 
        print('Before sorted: \n  {}\n  {}'.format(classical_eps, classical_results))

        sorted_idx = np.argsort(classical_eps)
        classical_eps = classical_eps[sorted_idx]
        classical_results = classical_results[sorted_idx]
        #classical_seed = classical_seed[sorted_idx]
        print('After sorted: \n  {}\n  {}'.format(classical_eps, classical_results))

        classical_eps_all.append(classical_eps)
        classical_results_all.append(classical_results)
classical_eps_all = np.array(classical_eps_all)
classical_results_all = np.array(classical_results_all) # (N_tests, eps_points)
```

```python
quantum_eps_all = []
quantum_results_all = []
for seed in seed_dir: 
    PATE_result_dir = os.path.join(seed, 'qPATE_epsilon.csv')
    if os.path.exists(PATE_result_dir):
        
        PATE_result = pd.read_csv(PATE_result_dir)
        quantum_seed = []
        quantum_eps = [] 
        quantum_results = [] 
        for i, k in enumerate(PATE_result.columns): 
            eps = float(float(k.replace(' - test_student_acc', '').split('_')[2].replace('e','')))
            test_student_acc = PATE_result[k].values[cutoff_epoch-1]
            quantum_eps.append(eps)
            quantum_results.append(test_student_acc)
            #quantum_seed.append(float(k.replace(' - test_student_acc', '').split('_')[0].replace('s',''))) 
            

        quantum_eps = np.array(quantum_eps)
        quantum_results = np.array(quantum_results)
        print('Before sorted: \n  {}\n  {}'.format(quantum_eps, quantum_results))

        sorted_idx = np.argsort(quantum_eps)
        quantum_eps = quantum_eps[sorted_idx]
        quantum_results = quantum_results[sorted_idx]
        print('After sorted: \n  {}\n  {}'.format(quantum_eps, quantum_results))

        quantum_eps_all.append(quantum_eps)
        quantum_results_all.append(quantum_results)

quantum_eps_all = np.array(quantum_eps_all)
quantum_results_all = np.array(quantum_results_all) # (N_tests, eps_points)
```

```python
cPATE_df = pd.DataFrame({'test_acc':classical_results_all.flatten(), 
                         'epsilon': classical_eps_all.flatten(), 
                         'pate': np.array(['Classical PATE' for _ in range(classical_results_all.flatten().shape[0])])})
qPATE_df = pd.DataFrame({'test_acc':quantum_results_all.flatten(), 
                         'epsilon': quantum_eps_all.flatten(), 
                         'pate': np.array(['Quantum PATE' for _ in range(quantum_results_all.flatten().shape[0])])})
df_all = pd.concat([cPATE_df, qPATE_df])
```

```python
plt.figure(figsize=(12,10))
#sns.violinplot(data=df_all,  x='epsilon', y='test_acc', hue='pate', cut=0)
sns.barplot(data=df_all,  x='epsilon', y='test_acc', hue='pate',errorbar="sd")
plt.xlabel('Epsilon', fontsize=21)
plt.ylabel('Student Accuracy (%)', fontsize=21)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.legend(fontsize=20, loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('MNIST Accuracy vs. Epsilon for Classical NN and VQC\n', fontsize=30)
plt.savefig('/Users/wangheehwan/Desktop/qPATE_lightning/Acc_Epsilon_epoch{}.pdf'.format(cutoff_epoch), bbox_inches='tight', dpi=300, format='pdf')
plt.show()
```

```python
quantum_teachers_all = []
quantum_results_all = []
for seed in seed_dir: 
    PATE_result_dir = os.path.join(seed, 'qPATE_teachers.csv')
    if os.path.exists(PATE_result_dir):
        
        PATE_result = pd.read_csv(PATE_result_dir)
        #quantum_seed = []
        quantum_teachers = [] 
        quantum_results = [] 
        for i, k in enumerate(PATE_result.columns): 
            teachers = int(k.replace(' - test_student_acc', '').split('_')[1].replace('t',''))
            test_student_acc = PATE_result[k].values[cutoff_epoch-1]
            quantum_teachers.append(teachers)
            quantum_results.append(test_student_acc)
            #quantum_seed.append(float(k.replace(' - test_student_acc', '').split('_')[0].replace('s',''))) 
            

        quantum_teachers = np.array(quantum_teachers)
        quantum_results = np.array(quantum_results)
        #quantum_seed = np.array(quantum_seed) 
        print('Before sorted: \n  {}\n  {}'.format(quantum_teachers, quantum_results))

        sorted_idx = np.argsort(quantum_teachers)
        quantum_teachers = quantum_teachers[sorted_idx]
        quantum_results = quantum_results[sorted_idx]
        #quantum_seed = quantum_seed[sorted_idx]
        print('After sorted: \n  {}\n  {}'.format(quantum_teachers, quantum_results))

        quantum_teachers_all.append(quantum_teachers)
        quantum_results_all.append(quantum_results)
quantum_teachers_all = np.array(quantum_teachers_all)
quantum_results_all = np.array(quantum_results_all) # (N_tests, teachers_points)
```

```python
cPATE_df = pd.DataFrame({'test_acc':classical_results_all.flatten(), 
                         'teachers': classical_teachers_all.flatten(), 
                         'pate': np.array(['classical PATE' for _ in range(classical_results_all.flatten().shape[0])])})
qPATE_df = pd.DataFrame({'test_acc':quantum_results_all.flatten(), 
                         'teachers': quantum_teachers_all.flatten(), 
                         'pate': np.array(['quantum PATE' for _ in range(quantum_results_all.flatten().shape[0])])})
df_all = pd.concat([cPATE_df, qPATE_df])
df_all = df_all.astype({'teachers':'int'})
```

```python
plt.figure(figsize=(12,10))
sns.barplot(data=df_all,  x='teachers', y='test_acc', hue='pate',errorbar="sd")
plt.xlabel('The Number of Teachers', fontsize=21)
plt.ylabel('Student Accuracy (%)', fontsize=21)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.legend(fontsize=20, loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('MNIST Accuracy vs. The Number of Teachers \nfor Classical NN and VQC\n', fontsize=30)
plt.savefig('/Users/wangheehwan/Desktop/qPATE_lightning/Acc_Teachers_epoch{}.pdf'.format(cutoff_epoch),  bbox_inches='tight', dpi=300, format='pdf')
plt.show()
```
